## Analysing data.csv

```python
import csv
from collections  import Counter
```

```python
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
```

```python
languages = []
popularity = []

for item in language_counter.most_common(15):   #gives 15 most common languages
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print(popularity)
```

```python
plt.barh(languages,popularity)        #barh - horizontal bar graph

plt.title("Most Popular Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()

plt.show()
```

```python
languages.reverse()
popularity.reverse()

plt.barh(languages,popularity)        #barh - horizontal bar graph

plt.title("Most Popular Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()

plt.show()
```
